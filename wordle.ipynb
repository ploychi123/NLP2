```python
words = []
for file_id in file_ids:
    words.append(nltk.corpus.webtext.words(file_id))
    
```

```python
words
```

```python
def preprocess_nested_words(list_of_lists):
    # สร้างลิสต์ว่างเพื่อเก็บคำที่ผ่านการ preprocess
    processed_words = []

    # วนลูปผ่านลิสต์ของลิสต์
    for word_list in list_of_lists:
        
        # วนลูปผ่านคำในลิสต์ปัจจุบัน
        processed_word_list = []
        for word in word_list:
            if word.isalpha() and len(word) >= 4:
                processed_word_list.append(word.lower())
        
        # เพิ่มลิสต์ของคำที่ผ่านการ preprocess ลงในลิสต์หลัก
        processed_words.append(processed_word_list)

    return processed_words



# เรียกใช้ฟังก์ชัน preprocess_nested_words
processed_data = preprocess_nested_words(words)
```

```python
processed_data
```

```python
list_word = []

for words in processed_data:
    for word in words:
        list_word.append(word)
```

```python
list_word
```

```python
# นับความถี่ของคำศัพท์ตามจำนวนตัวอักษร
word_frequency = {}

for word in list_word:
    length = len(word)
    if length not in word_frequency:
        word_frequency[length] = 1
    else:
        word_frequency[length] += 1


df = pd.DataFrame(list(word_frequency.items()), columns=['Word Length', 'Frequency'])
df_sorted = df.sort_values(by='Word Length')
print(df_sorted)
```
