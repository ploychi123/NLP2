### Web scraping

เป็นวิธีการในการดึงข้อมูลจากเว็บไซต์/เว็บเพจ โดยเขียนโปรแกรมเพื่อเลือกดึงข้อมูลที่ต้องการมาอัตโนมัติ แล้วนำไปจัดรูปแบบให้พร้อมนำไปใช้งานต่อไป

#### วิธีการดึงข้อมูลจากเว็บไซต์ 
- เรียกใช้ API (เว็บไซต์ต้องเปิดให้ใช้งานได้)
- ใช้ฟังก์ชัน google sheet (เฉพาะที่เป็น table/list)
- ทำ web scraping (เว็บไซต์อนุญาตให้ดึงข้อมูล)

#### ประโยชน์ ?
#### ข้อจำกัด ?

#### Library ที่จำเป็นต้องใช้
- requests ดึง source (html) การแสดงผลในหน้าเว็บที่ต้องการมาเก็บไว้ในรูปแบบ text 
- BeautifulSoup ดึง/สกัดข้อมูลที่ต้องการจาก source ที่เก็บอยู่ในรูป html หรือ xml
- csv จัดเก็บข้อมูลที่สกัดได้ เพื่อนำไปใช้งานต่อไป

#### จริยธรรมในการดึงข้อมูล
- ตรวจสอบได้โดยเติม /robots.txt หลัง url

#### ขั้นตอนการทำ web scraping
1. เขียนคำสั่งเพื่อส่ง request ไปยัง URL ของเว็บเพจที่ต้องการดึงข้อมูล
2. รับ response ที่ server ตอบกลับมาในรูปแบบ html หรือรูปแบบอื่นๆ
3. ตรวจสอบว่า response เป็น source ของ html หน้าที่ต้อง (request สำเร็จ status = 200)
4. เปลี่ยน source ที่ได้ให้อยู่ใน class BeautifulSoup เพื่อเรียกใช้ method ในการดึงข้อมูลได้
5. เลือก tag ที่ต้องการดึงข้อมูล แล้วสกัดข้อมูลที่เก็บอยู่ใน tag นั้นออก
6. จัดเก็บข้อมูลที่ได้ในรูปแบบที่มีโครงสร้าง เช่น csv

```python
import requests
from bs4 import BeautifulSoup
import csv
import re
```

```python
url = "https://www.azlyrics.com/lyrics/bangtanboys/springday.html"
#url = "https://www.azlyrics.com/lyrics/nct/candy.html"
```

```python
response = requests.get(url)
#response.encoding = "utf-8"
print(type(response))
print(response)

if response.status_code == 200:
    print("success")
else:
    print('not success')
```

```python
html = BeautifulSoup(response.text, 'html.parser')
print(type(html))
print(html)
```

```python
print(html.prettify())
```

```python
print(html.title.string)
```

```python
h2 = html.find_all('h2')
print(h2)

for h in h2:
    print(h.text)
```

```python
#body > div.container.main-page > div.row > div.col-xs-12.col-lg-8.text-center

selector = 'div.col-xs-12.col-lg-8.text-center'
lyrics = html.select_one(selector)
print(lyrics)
```

```python
print(lyrics.text)
```

```python
#strip() ตัด space ข้างบนและข้างล่าง
lyrics_text =lyrics.text.strip()
print(type(lyrics_text))
print(lyrics_text)
```

```python
end_pos = lyrics_text.find("Submit Corrections")
lyrics_text = lyrics_text[:end_pos]
print(lyrics_text.strip())
```

```python
lst_div = lyrics.find_all('div')
for i, div in enumerate(lst_div):
    print(i, div.text)
```

```python
song = lst_div[2].text
name = lst_div[3].text
lyric = lst_div[5].text
print("Song:", song.replace("lyrics",""))
print("Artist:", name.strip().replace("Lyrics",""))
print("Lyric:")
print(lyric.strip())
```
