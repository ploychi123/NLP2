```python
frequency_table = {}

for word in word_len4:
    length = len(word)
    if length in frequency_table:
        frequency_table[length] += 1
    else:
        frequency_table[length] = 1

df = pd.DataFrame(list(frequency_table.items()), columns=['ความยาว', 'จำนวนคำ'])
```

```python
def user_input():
  return str(input("กรุณาใส่คำที่ต้องการทายความยาว {0} ตัวอักษร : ".format(desired_length))).lower()

def check_guess(secret_word, guess, desired_length):
    pattern = re.compile(f"^[a-z]{{{desired_length}}}$")
    if not pattern.match(guess):
        return False, f"ความยาวตัวอักษรไม่ถูกต้อง"

    correct_positions = [char1 for char1, char2 in zip(secret_word, guess) if char1 == char2]

    common_characters = [s for s in secret_word if s in guess]
    filtered_characters = [set(s for s in common_characters if s not in correct_positions)]

    missing_characters = [set(s for s in guess if s not in secret_word)]

    feedback = f"คำทาย \"{guess}\": ตำแหน่งตัวอักษร {correct_positions} ถูกต้อง, มี {filtered_characters} ที่อยู่ในคำที่ถูกต้อง, ไม่มีตัวอักษร {missing_characters} ในคำศัพท์"

    if len(correct_positions) == desired_length:
        print(f"คำถูกต้อง! {secret_word}")
        return True
    else:
        return False, feedback
```

```python
from nltk.corpus import wordnet
def get_word_meaning(word):
    synsets = wordnet.synsets(word)

    if synsets:
        # ใช้ synsets[0] เพื่อให้แสดงความหมายของคำที่มีประโยชน์ที่สุด
        print(f'Meaning of "{word}": {synsets[0].definition()}')
    else:
        print(f'Meaning of "{word}": not found')
```

```python
desired_length = int(input("กรุณาใส่ความยาวของคำที่ต้องการสุ่ม : "))
random_word = random.choice([word for word in word_len4 if len(word) == desired_length]).lower()
print(random_word)
```

```python
loops = 0
while loops < 5:
  word = user_input()
  get_word_meaning(word)

  if check_guess(random_word, word, desired_length) == True:
    break
  else:
    print(check_guess(random_word, word, desired_length))
  print("")

  loops += 1
  if loops == 5:
    print("")
    print("หมดเวลา")
    print(f"คำที่ถูกต้องคือ : {random_word}")
    get_word_meaning(random_word)
```
