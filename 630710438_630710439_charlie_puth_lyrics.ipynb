```python
import requests    # request data from web
import bs4         # access data from web
import pandas as pd
import time
```

```python
url = "https://www.azlyrics.com/c/charlieputh.html?fbclid=IwAR0JSn4zCJIAODzVHRhDIUrZ4nJP9YpPRUXnHUzYXNx-G2BHdcV2mBDd2ew"
source = requests.get(url)
```

```python
# แสดงชนิดข้อมูลที่ได้
print(type(source))

# แสดงสถานะของ request 
print(source)
```

```python
# แสดง source code ที่ได้จาก method get
print(source.text)
```

```python
html = bs4.BeautifulSoup(source.text, "html.parser")
print(type(html))
```

```python
print(html)
```

```python
links = html.findAll('div', class_="listalbum-item")
links
```

```python
links[0]
```

```python
link_texts = [a_tag.find('a')['href'] for a_tag in links]

# หรือถ้าต้องการแสดงผลลัพธ์
for link_text in link_texts:
    if link_texts:
        print(link_text)
```

```python
linkfull = []

for link in link_texts:
    if link.startswith("https://www.azlyrics.com"):
        # ถ้าลิงค์มี https://www.azlyrics.com แล้วให้ไปเก็บไว้ใน linkfull
        linkfull.append(link)
    else:
        # ถ้าไม่มี https://www.azlyrics.com ให้เติมข้างหน้าลิงค์ด้วย https://www.azlyrics.com และเก็บไว้ใน linkfull
        linkfull.append("https://www.azlyrics.com" + link)
    
print(len(linkfull))
# พิมพ์ linkfull
for link in linkfull:
    print(link)
```

```python
artist = []
song = []
lyrics = []

selector = 'div.col-xs-12.col-lg-8.text-center'

for link in linkfull:

    music1_html = bs4.BeautifulSoup(requests.get(link).text, "html.parser")
    lyrics_scrap = music1_html.select_one(selector)
    lst_div = lyrics_scrap.find_all('div')
    song.append(lst_div[2].text)
    artist.append(lst_div[3].text)
    lyrics.append(lst_div[5].text)


    time.sleep(5)
```

```python
print(artist)
print(lyrics)
print(song)
```

```python
data = {
    'Artist': artist,
    'Lyrics': lyrics,
    'Song': song,
}

# Create a DataFrame from the dictionary
df = pd.DataFrame(data)

df
```

```python
# Clean the 'Artist' column
df['Artist'] = df['Artist'].str.replace(r'\n|Lyrics', '', regex=True)
df['Lyrics'] = df['Lyrics'].str.replace(r'\n|\r', ' ', regex=True)
df['Song'] = df['Song'].str.replace(r'"|lyrics', '', regex=True)
df
```

```python
# Save the DataFrame to a CSV file
df.to_csv('charlie_puth.csv', index=False)
```
